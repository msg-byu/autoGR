```python
def get_n_vals(n):
    first = int(n**(1/3.))
    a = None
    b = None
    for i in range(int(n**(1/3.)+1)):
        for m in [1,2,4]:
            temp = m*i**3
            if temp != first**3:
                if temp < n:
                    if a is None or n-a>n-temp:
                        a = temp
                else:
                    if b is None or b-n>temp-n:
                        b = temp
    return([a,first**3,b])
```
