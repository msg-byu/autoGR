```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def primes(n):
    primfac = {}
    d = 2
    while d*d <= n:
        while (n % d) == 0:
            if d not in primfac.keys():
                primfac[d] =1  # supposing you want multiple factors repeated
            else: 
                primfac[d] += 1
            n //= d
        d += 1
    if n > 1:
       primfac[n] = 1
    return primfac
```

```python
def n_HNFs(n):
    total = 1
    ps = primes(n)
    for k in ps:
        temp = (k**(ps[k]+2)-1)*(k**(ps[k]+1)-1)/((k+1)*(k-1)**2)
        total *= temp
    
    return total       
```

```python
vals = []
for i in range(1000):
    vals.append(n_HNFs(i))
```

```python
plt.plot(vals)
plt.title("Number of Supercells for a Volume")
plt.xlabel("Volume Factor")
plt.ylabel("Number of Supercells")
plt.loglog()
plt.show()
```

```python
tols = [1]
for i in range(2,100):
    tols.append(tols[-1]+n_HNFs(i))
```

```python
plt.plot(tols)
plt.title("Cumulative Number of Supercells Over Volumes",fontsize=16)
plt.xlabel("Volume Factor",fontsize=14)
plt.ylabel("Cumulative Number of Supercells",fontsize=12)
plt.loglog()
plt.savefig('/Users/wileymorgan/Desktop/N_sups.pdf')
plt.show()
```
